### 0 - Import libraries or dependencies

```python
import json
from gitsource import GithubRepositoryDataReader
```

### 1 - Download files to homework from git

```python
def fetch_documents():
    """Fetch all .md lesson files from the llm-zoomcamp repository."""
    reader = GithubRepositoryDataReader(
        repo_owner="DataTalksClub",
        repo_name="llm-zoomcamp",
        commit_id="8c1834d",
        allowed_extensions={"md"},
        filename_filter=lambda path: "/lessons/" in path,
    )
    files = reader.read()
    return [file.parse() for file in files]
```

### 2 - Each file has a parse() method that returns a dictionary with its filename and content:


```python
documents = []

for file in files:
    doc = file.parse()
    documents.append(doc)
```

### 3 - Save documents to disk so the next notebook can reuse them

```python
import json

# Persist the parsed documents to a JSON file.
# 02-index-search.ipynb will load this file instead of
# fetching everything from GitHub again.
with open("documents.json", "w", encoding="utf-8") as f:
    json.dump(documents, f, ensure_ascii=False, indent=2)

print(f"Saved {len(documents)} documents to documents.json")
```

```python
print(f"They have {len(documents)} documents in documents.json")
```
