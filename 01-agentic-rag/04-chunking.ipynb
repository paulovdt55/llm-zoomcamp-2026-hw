# 04 - Chunking (Q4)


### 1 - Load the documents produced 

```python
import json


def load_documents(path="documents.json"):
    """Load the documents produced by 01-fetch-files.ipynb."""
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)


documents = load_documents()
len(documents)  # should print 72
```

### 2 - Split the documents into overlapping chunks

A 2000-char window slides forward in 1000-char steps, so consecutive chunks overlap by 1000 characters.

```python
from gitsource import chunk_documents

# size = window width (characters per chunk)
# step = how far the window moves each time (overlap = size - step = 1000)
chunks = chunk_documents(documents, size=2000, step=1000)
```

### 3 - Inspect one chunk

Confirm the new fields: each chunk keeps `filename` and adds `start` and `content`.

```python
chunks[0]
```

### 4 - Answer to Q4: how many chunks?

```python
len(chunks)
```

### 5 - Save the chunks for Q5



```python
with open("chunks.json", "w", encoding="utf-8") as f:
    json.dump(chunks, f, ensure_ascii=False, indent=2)

print(f"Saved {len(chunks)} chunks to chunks.json")
```
